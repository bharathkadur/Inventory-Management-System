```python
import json
z = {1:{"name":"jam", "quantity":30, "price":20},
     2:{"name":"Bread", "quantity":50, "price":40},
     3:{"name":"Butter", "quantity":10, "price":100},
     4:{"name":"Cheese", "quantity":50, "price":50},
     5:{"name":"peanut butter", "quantity":40, "price":20},
     6:{"name":"Chiken", "quantity":40, "price":50},
     7:{"name":"EGG", "quantity":50, "price":70},
     8:{"name":"Wheat Flour", "quantity":20, "price":100},
     9:{"name":"Fish", "quantity":70, "price":10},
     10:{"name":"Cello Blue ink pen", "quantity":6000, "price":5},
     11:{"name":"Apsara Pencil", "quantity":500, "price":40},
     12:{"name":"Natraja Pencil", "quantity":800, "price":60},
     13:{"name":"Dog Biscuit", "quantity":40, "price":20},
     14:{"name":"Motorola Mobile E7", "quantity":1, "price":10000},
     15:{"name":"Oppo K3", "quantity":50, "price":50000},
     16:{"name":"Dairy Milk", "quantity":100, "price":10},
     17:{"name":"Dairy Milk Silk", "quantity":500, "price":100},
     18:{"name":"Hide and seek Coffe Flavour", "quantity":50, "price":40},
     19:{"name":"Cheetos", "quantity":50, "price":500},
     20:{"name":"Mentos Mix Flavour", "quantity":50, "price":10},
    }

fd=open('items.json','w')
c=json.dumps(z,indent=4)
fd.write(c)
fd.close()

x = {1:{"name":"jam", "quantity":30, "price":20},
     2:{"name":"Bread", "quantity":50, "price":40},
     3:{"name":"Butter", "quantity":10, "price":100},
     4:{"name":"Cheese", "quantity":50, "price":50},
     5:{"name":"peanut butter", "quantity":40, "price":20}
    }

def Read():
    fd=open('inventory.json','r')
    global y
    y=fd.read()
    x=json.loads(y)
    fd.close()
    
def Read1():
    fd=open('items.json','r')
    global b
    b=fd.read()
    z=json.loads(b)
    fd.close()
    
def Display1():
    Read1()
    for i in z.keys():
        print("id="+str(i))
        print(z[i])
            

def Display():
    Read()
    for i in x.keys():
        print("id="+str(i))
        print(x[i])
        
def Add():
    Display1()
    print("\nSelect the product from the above list\n")
    print("Enter the id")
    id=int(input())
    a={"name":input("Enter name, quantity, price "), "quantity":int(input()), "price":int(input())}
    x[id]=a
    Update()

def Update():
    fd=open('inventory.json','w')
    y=json.dumps(x,indent=4)
    fd.write(y)
    fd.close()

def Purchase():
    item=input("Enter the item name and quantity ")
    q=int(input())
    f=0
    Read()
    for i in range(len(x)):
        id=i+1
        if x[id]['name']==item:
            f=1
            if x[id]['quantity']>=q:
                x[id]['quantity']-=q
                Update()
                print("Cost = "+str(x[id]['price']*q)+"\n")
            else:
                print("Not Available")
                print(str(x[id]['quantity'])+" are available.\n")
    if f==0:
        print("Invalid item")
        
def MaxAmount():
    Read()
    item = 0 
    amount = 0
    for i in range(len(x)):
        id=i+1
        amount+=(x[id]['price']*x[id]['quantity'])
        item+=x[id]['quantity']
    print("Total no. of items available = "+str(item)+"\nTotal Cost = "+str(amount)+"\n")
                 
Update()
while(1):
    print("Select the function to perform \n 1.Read the contents \n 2.Add a new item \n 3.Purchase an item \n 4.Display Total No. of Items and Total Cost \n 5.Display items.json \n press any key to exit ")
    n=int(input())

    if n==1:
        Display()
    elif n==2:
        Add()
    elif n==3:
        Purchase()
    elif n==4:
        MaxAmount()
    elif n==5:
        Display1()
    else:
        break
```
